```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.covariance import EllipticEnvelope

# Load dataset
data = pd.read_csv('cleaned_sample.csv')

# Select numerical columns for PCA
numerical_columns = data.select_dtypes(include=[np.number]).columns.tolist()
numerical_columns.remove('ID')  # Exclude 'ID' column

# Separate features and target variable
X = data[numerical_columns]
y = data['Diagnosis']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Score plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='coolwarm', alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Score Plot')
plt.legend(loc='upper right')
plt.show()

# Outlier plot
plt.figure(figsize=(10, 6))
env = EllipticEnvelope(contamination=0.1)  # Adjust contamination as needed
env.fit(X_scaled)
outliers = env.predict(X_scaled) == -1
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=outliers, palette='coolwarm', alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Outlier Plot')
plt.legend(loc='upper right', labels=['Inliers', 'Outliers'])
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load dataset
data = pd.read_csv('cleaned_sample.csv')

# Categorize concavity levels into low, medium, and high (You may need to adjust these thresholds based on your data)
data['Concavity_Level'] = pd.cut(data['concavity1'], bins=[0, 0.1, 0.5, float('inf')], labels=['Low', 'Medium', 'High'])

# Construct contingency table
contingency_table = pd.crosstab(data['Diagnosis'], data['Concavity_Level'])

# Perform chi-square test
chi2_stat, p_val, _, _ = chi2_contingency(contingency_table)
print(chi2_stat, p_val)

# Interpret the results
if p_val < 0.05:
    print("The chi-square test yielded a significant association between diagnosis and concavity levels.")
    print("Concavity could serve as a relevant factor in distinguishing between malignant and benign tumor cells.")
else:
    print("No significant association was found between diagnosis and concavity levels.")
```
